```python
import pandas
import matplotlib.pyplot as plt

global dataset
dataset = pandas.read_csv('BSB Fundraiser Progress.csv',parse_dates=['Datestamp (sampled at 10:30 pm PST every day)'], infer_datetime_format=True)

plt.figure(figsize=(16,10), dpi=100)

plt.gca().set(title="Funding Progress (Percentage)", xlabel='Date', ylabel="Percent Funded")
plt.plot(dataset['Datestamp (sampled at 10:30 pm PST every day)'], dataset['Percent funded'], color='tab:red')

plt.show()
```

```python
plt.figure(figsize=(16,10), dpi=100)

plt.gca().set(title="Backer Count Progression", xlabel='Date', ylabel="Backer Count")
plt.plot(dataset['Datestamp (sampled at 10:30 pm PST every day)'], dataset['Backer count'], color='tab:red')

plt.show()
```

```python
plt.figure(figsize=(16,10), dpi=100)

dataset['Datestamp (sampled at 10:30 pm PST every day)'] = pandas.to_datetime(dataset['Datestamp (sampled at 10:30 pm PST every day)'])

global differential
differential = dataset.diff()

plt.gca().set(title="Daily Change in Backer Count", xlabel='Date', ylabel="Backer Count Difference")
plt.bar(dataset['Datestamp (sampled at 10:30 pm PST every day)'], differential['Backer count'], color='tab:blue')

plt.show()
```

```python
plt.figure(figsize=(16,10), dpi=100)

plt.gca().set(title="Daily Change in Funding Percentage", xlabel='Date', ylabel="Funding Percentage Difference")
plt.bar(dataset['Datestamp (sampled at 10:30 pm PST every day)'], differential['Percent funded'], color='tab:blue')

plt.show()
```
